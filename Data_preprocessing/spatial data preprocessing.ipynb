```python
# 자전거 거치소 데이터

import pandas as pd

bike_stations = pd.read_csv('자전거대여소.csv', encoding='utf-8')
bike_stations.sort_values(by = '대여소 번호', ascending = True)
```

```python
# 자전거 거치소별로 정리하고 싶은 데이터프레임 읽기
# 원하는 데이터프레임으로 변경

culture_df = pd.read_csv('Culture.csv', encoding='cp949')
culture_df.head()
```

```python
# 이 때 각 데이터프레임 내에 특이한 값들이 들어가 있으니 사전에 검토해주시는 게 좋아요!

# 1. 위도, 경도에 nan 값이 포함되어 있는지 확인
    # 포함되어 있다면 제거 -> 너무 많이 nan값이라면 다른 방법으로 위도, 경도 정보 받아오기
    
culture_df.isna().sum() # 위도, 경도 결측치 3개
```

```python
# 위에서 찾은 결측치 제거하기
culture_df = culture_df.dropna(subset=['위도'])
culture_df.isna().sum()
```

```python
# 2. 각 columns 별 describe()를 찍어서 min, max 값 확인하기
    # 종종 위도, 경도가 한개씩 바껴있는 게 있더라구요..!
print(culture_df['위도'].describe())
print()
print(culture_df['경도'].describe())
```

```python
# 3. 위도, 경도의 columns의 형태 확인
    # float 형식이어야 하기 때문에 info()를 찍어서 어떤 type인지 확인합니다
    # 만약 string type이라면
    # 데이터프레임명[컬럼명] = 데이터프레임명[컬럼명].astype(float) 로 변경해주시면 됩니다
# 위도, 경도 float type으로 변환
culture_df['위도'] = culture_df['위도'].astype('float')
culture_df['경도'] = culture_df['경도'].astype('float')

culture_df.info()
```

```python
# 이미 해당하는 데이터에 위도와 경도가 포함되어 있는 경우
# 위도, 경도 columns을 활용해서 각 대여소별 500m 반경으로 정리하기

# goal_df: 500m 내로 정리하고 싶은 목표 데이터프레임
# station: 반복문을 통해서 들어오는 station의 정보 -> 위도, 경도를 확인
# radius: 원하는 반지름으로 설정

from geopy.distance import geodesic

def count_within_radius(goal_df, station, radius):
    
    # 반경 500m 이내에 있는 문화시설 개수 파악
    count = 0
    
    for i in goal_df.index:
        
        # 문화시설과 자전거 대여소 간의 거리 계산
        distance = geodesic(
            
            # 이 때, goal_df에 위도, 경도가 어떤 컬럼명으로 설정되어있는지 확인하고 변경해주셔야 합니다
            (goal_df.at[i, "위도"], goal_df.at[i, "경도"]), 
            (station['위도'], station["경도"])
        ).meters
        
        # 반경 내에 있는 문화시설인지 확인
        if distance <= radius:
            count += 1
    
    return count

# 각 자전거 대여소별 반경 500m 이내의 문화시설 수
# 하나의 데이터프레임을 만들어서 내부적으로 데이터를 저장합니다
bike_culture = pd.DataFrame(columns=['자전거대여소 번호', '인근 문화시설 수'])

# 각 자전거 대여소별로 반경 500m 이내의 문화시설 개수 파악
# 반복문에서 station_name과 station 파라미터가 bike_stations의 한 줄로 들어간다고 생각해주세요!

for i in bike_stations.index:
    station_name = str(bike_stations.at[i, "대여소 번호"])
    culture_count = count_within_radius(culture_df, bike_stations.loc[i], 500)
    
    bike_culture.loc[i] = [station_name, culture_count]
    
    # 잘 돌아가고 있는지 확인하기 위한 프린트 구문
    print(i)

# 각 자전거대여소별 500m 이내의 문화시설 개수를 포함하고 있는 데이터프레임이 생성됩니다
bike_culture
```

```python
# university_df를 다룰 때 폐교와 관련된 정보를 전처리하여 제거하였으나
# 코드를 정리하고 있기 때문에 여기서는 따로 전처리를 제외하겠습니다!

# 대학교의 경우 위도, 경도에 대한 정보가 존재하지 않아서
# 도로명 주소를 사용하고자 하였으나 도로명 주소에 nan 값이 너무 많이 존재하여
# 각 대학교명을 가지고 google api를 사용해서 위도, 경도 정보를 받아왔습니다

university_list = university_df['학교명']
university_list
```

```python
import requests

university_x = []
university_y = []

def get_geocode(address, api_key):
    url = f"https://maps.googleapis.com/maps/api/geocode/json?address={address}&key={api_key}"
    response = requests.get(url)
    data = response.json()
    
    if data['status'] == 'OK':
        location = data['results'][0]['geometry']['location']
        latitude = location['lat']
        longitude = location['lng']
        return latitude, longitude
    else:
        return None, None

# API 키 설정
api_key = 'AIzaSyCiZYyFONuYFzEm8cxdCxjJB-u1fN2udSI'

# 주소 설정
# 좌표 가져오기
for address in university_list:
    latitude, longitude = get_geocode(address, api_key)

    if latitude and longitude:
        university_x.append(latitude)
        university_y.append(longitude)

    print(address)
```

```python
# 위의 과정을 통해 위도, 경도에 대한 정보를 모두 받아오게 되었다면
# 리스트 형태를 가지고 반경의 정보를 받을 수 있도록
# 새로운 함수를 하나 정의합니다
# 위에서 정의했떤 500m 이내의 문화시설 찾기와 매우 비슷하지만
# 데이터프레임과 리스트 형태의 차이가 있어 따로 분류하여 정의하였습니다

def count_within_radius_using_list(x_list, y_list, station, radius):
    
    # 반경 500m 이내에 있는 대학교 개수 파악
    count = 0
    
    for i in range(len(university_x)):
        
        # 지하철역과 자전거 대여소 간의 거리 계산
        distance = geodesic(
            (x_list[i], y_list[i]),
            (station['위도'], station["경도"])
        ).meters
        
        # 반경 내에 있는 지하철역인지 확인
        if distance <= radius:
            count += 1
    
    return count

# 각 자전거 대여소별 반경 500m 이내의 지하철역 개수 저장 데이터프레임
bike_university = pd.DataFrame(columns=['자전거대여소 번호', '인근 대학교 개수'])

# 각 자전거 대여소별로 반경 500m 이내의 대학교 개수 파악
for i in bike_stations.index:
    station_name = str(bike_stations.at[i, "대여소 번호"])
    
    # 500m로 정리되기 원하는 정보가 리스트 형태라면
    # 위도(x), 경도(y)를 각각의 파라미터로 넘겨줍니다
    university_count = count_within_radius_using_list(university_x, university_y, bike_stations.loc[i], 500)
    
    bike_university.loc[i] = [station_name, university_count]
    
    print(i)

bike_university
```
